```python
import os
import jieba
import re
from chardet import detect
import itertools
```

```python
def get_encoding_type(file):
    with open(file, 'rb') as f:
        rawdata = f.read()
    return detect(rawdata)['encoding']
```

```python
tarfile= 'temp.txt'
file_n='NBA-1.txt'
codec=get_encoding_type(file_n)
with open(file_n, 'r', encoding=codec,errors='ignore') as f, open(tarfile, 'w', encoding='utf-8',errors='replace') as e:
    text = f.read()
    e.write(text)
path = "./"+file_n
os.remove(path)
os.rename(tarfile,file_n)
```

```python
with open(file_n, 'r', encoding='utf-8') as f:
    lines=f.readlines()
with open('NBA_en.txt','r', encoding='utf-8') as t:
    tl_lines=t.readlines()


corpous=[]
for line,tl_line in zip(lines,tl_lines):
    seg_list  = jieba.cut(line , cut_all = False )
    st='/'.join(seg_list)
    eng_wrd = tl_line.split(' ')
    chn_wrd = st.split('/')
    if '：' in chn_wrd:
        chn_wrd.remove('：')
    if ':' in chn_wrd:
        chn_wrd.remove(':')
    corpous.append([chn_wrd,eng_wrd])
```

```python
corpus=dict()

for item in corpous:
    for wrd in item[0]:
        corpus[wrd]=[]

for item in corpous:
    for w in range(len(item[0])):
        try:
            corpus[item[0][w]].append(item[1][w])
        except:
            continue
```

```python
for key in corpus:
    if len(corpus[key])<10:
        print(key, corpus[key])
```
